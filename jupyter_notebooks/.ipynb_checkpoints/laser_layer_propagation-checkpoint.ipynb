```python
import numpy as np
from scipy.ndimage import rotate
import pyvista
```

```python
# matrix

def makeGaussian(size, fwhm=3, center=None, amp=1):
    """ Make a square gaussian kernel.

    size is the length of a side of the square
    fwhm is full-width-half-maximum, which
    can be thought of as an effective radius.
    """

    x = np.arange(0, size, 1, float)
    y = x[:, np.newaxis]

    if center is None:
        x0 = y0 = size // 2
    else:
        x0 = center[0]
        y0 = center[1]

    return np.exp(-4*np.log(2) * ((x-x0)**2 + (y-y0)**2) / fwhm**2)/ (2*np.pi*fwhm**2)/(-4*np.log(2))**2

def makeRod(size, width):
    arr = np.zeros([size]*2)
    arr[:, size//2-width//2 : size//2+width//2] = 1
    return arr
```

```python
size = (100, 80)
fwhm = 10

arr3d = np.array([makeGaussian(size[0], fwhm=fwhm)]*size[1])
plotter = pyvista.Plotter()
plotter.add_mesh(arr3d[:, :, :50], show_edges=False)
plotter.show()
```

```python
rot1 = rotate(arr3d, 20, mode='nearest', reshape=False)
plotter = pyvista.Plotter()
plotter.add_mesh(rot1[:, :, :50], show_edges=False)
plotter.show()
```

```python
lay1, lay2 = rot1, arr3d
rot2 = np.vstack([lay1, rotate(lay2,-15, mode='nearest', reshape=False)[size[1]//2:]])
print(lay1.shape)
print(np.where(lay1[len(lay1), :, :]))
print(np.where(rot2[len(rot2), :, :]))
                   
plotter = pyvista.Plotter()
plotter.add_mesh(rot2[:, :, :50], show_edges=False)
plotter.show_grid()
plotter.show()
```
