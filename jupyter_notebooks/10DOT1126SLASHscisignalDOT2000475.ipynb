```python
from sympy import *
```

```python
nhp, nhnp, nhprot, nlp, nlnp, nlprot = symbols("N_H^P N_H^NP N_H^PROT N_L^P N_L^NP N_L^PROT ")
a, b, x, y, z = symbols("a b x y z")
a_ = Eq(a, nhp/nhnp)
b_ = Eq(b, nlp/nlnp)
cond_constraint = Eq((nhp+nhnp)/nhprot, (nlp+nlnp)/nlprot)

x_ = Eq(x, nhp/nlp)
y_= Eq(y, nhnp/nlnp)
z_ = Eq(z, nhprot/nlprot)
```

```python
system = [cond_constraint, x_, y_, z_, a_, b_]
sys = [s.lhs-s.rhs for s in system]

linsolve(sys, a)
```

```python
all_symbols = set()
for e in system:
    all_symbols.update(e.free_symbols)
all_symbols
```

```python
subbed_b = (cond_constraint
          .simplify()
          .expand()
          .subs([(nhprot, z*nlprot)])
          .simplify()
          .expand()
          .subs([(nlprot,nlprot/z)])
          .simplify()
         
          .subs([(nlprot, 1)])
         
          .subs([(nlp, nhp/x)])
          .subs([(nlnp, nhnp/y)])
          .simplify()
         
          .subs([(nhp, b*nhnp)])
          .subs([(nhnp, 1)])
         )
subbed_b
```

```python
subbed_a = (cond_constraint
          .simplify()
          .expand()
          .subs([(nhprot, z*nlprot)])
          .simplify()
          .expand()
          .subs([(nlprot,nlprot/z)])
          .simplify()
         
          .subs([(nlprot, 1)])

          .subs([(nhnp, nlnp*y)])
          .subs([(nhp, nlp*x)])

          .subs(nlnp, nlp/a)
          .subs(nlp, 1)
         )
subbed_a
```

```python
solve(subbed_a, a)[0]
```

```python
solve(subbed_b, b)[0]
```
