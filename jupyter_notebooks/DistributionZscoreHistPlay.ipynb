```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, center=0, sigma=0.5, amp=1):
    return amp*np.exp(-(x-center)**2 / 2 / sigma**2)

def zscore(x):
    return (x-np.mean(x))/np.std(x)

def hist_vect(x, nbins):
    pass
    
```

```python
x = np.linspace(-2, 2, 1000)
g1 = gauss(x, 1, 0.4, .5)
g2 = gauss(x, -1, 0.4, .5)
plt.plot(x,g1+g2)
plt.title("Bimodal distribution")
plt.ylim(0,1)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))

ax1.hist(zscore(g1+g2), bins=64, alpha=0.5, label='z-scored')
ax1.set_title("Bimodal distribution histogram")
ax1.legend()
ax2.hist(g1+g2, bins=64, alpha=0.5, label='raw')
ax2.set_title("Bimodal distribution histogram")
ax2.legend()

plt.show()
```

```python
g3 = gauss(x, 0, 0.4, 0.5)
plt.plot(x, g3)
plt.title("Unimodal distribution (sigma=0.4)")
plt.ylim(0,1)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))

ax1.hist(zscore(g3), bins=64, alpha=0.5, label='z-scored')
ax1.set_title("Unimodal distribution histogram (sigma=0.4)")
ax1.legend()
ax2.hist(g3, bins=64, alpha=0.5, label='raw')
ax2.set_title("Unimodal distribution histogram (sigma=0.4)")
ax2.legend()

plt.show()
```

```python
g4 = gauss(x, 0, 1, 0.5)
plt.plot(x, g4)
plt.title("Unimodal distribution (sigma=1)")
plt.ylim(0,1)
plt.show()
```
