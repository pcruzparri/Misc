```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def int_area(w1, w2, theta):
    return w1*w2/np.sin(theta)

def d1(w1, w2, theta):
    return np.sqrt(((w1/np.sin(theta)) + (w2/np.tan(theta)))**2 + w2**2)

def d2(w, theta):
    return w/np.sin(theta)
```

```python
w1 = 50 #um
w2 = 50 #um
angles = np.linspace(5, 25, 40)
int_areas = [int_area(w1,w2, np.deg2rad(angle)) for angle in angles]
dist1 = [d1(w1,w2, np.deg2rad(angle)) for angle in angles]
dist2 = [d2(w1, np.deg2rad(angle)) for angle in angles]
dist3 = [d2(w2, np.deg2rad(angle)) for angle in angles]

plt.plot(angles, int_areas)
plt.show()
```

```python
plt.plot(angles, dist1)
plt.plot(angles, dist2)
plt.plot(angles, dist3)
plt.show()
```
