```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R
from mpl_toolkits.mplot3d import axes3d
```

```python
x_rot = lambda theta: np.array([[1, 0, 0], 
                                [0, np.cos(theta), -np.sin(theta)], 
                                [0, np.sin(theta), np.cos(theta)]])
y_rot = lambda theta: np.array([[np.cos(theta), 0, np.sin(theta)],
                                [0, 1, 0], 
                                [-np.sin(theta), 0, np.cos(theta)]])
z_rot = lambda theta: np.array([[np.cos(theta), -np.sin(theta), 0], 
                                [np.sin(theta), np.cos(theta), 0],
                                [0, 0, 1]])
```

```python
l1 = np.array([1, 0, 0])

l1_new = y_rot(np.deg2rad(45))@l1
l1_new
```

```python
ax = plt.figure().add_subplot(projection='3d')
ax.plot(*[np.linspace(0, i) for i in l1])
ax.plot(*[np.linspace(0, i) for i in l1_new])
ax.legend()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_zlim(-5,5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
