```python
from koinapy import Koina
from time import perf_counter
import pandas as pd
from numpy.random import choice, randint, seed
import itertools as it
```

```python
canonical_amino_acids = 'ACDEFGHIKLMNPQRSTVWY'

def generate_random_peptide(length):
    return ''.join(choice(list(canonical_amino_acids), size=length))
```

```python
model_name = "pfly_2024_fine_tuned"
model = Koina(model_name)

max_peptide_length = 30
n_inputs = 2100000
peptides = [generate_random_peptide(randint(10, max_peptide_length + 1)) for _ in range(n_inputs)]

```

```python
# Calling all peptides
inputs = pd.DataFrame({'peptide_sequences': peptides})

start_time = perf_counter()
predictions = model.predict(inputs)
end_time = perf_counter()
print(f"Time taken for {n_inputs} predictions: {end_time - start_time:.2f} seconds")
```
